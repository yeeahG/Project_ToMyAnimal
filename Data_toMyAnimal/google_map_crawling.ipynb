```python
# locates = 'div > div > span'
locates = 'div.m6QErb.DxyBCb.kA9KIf.dS8AEf.ecceSd'
button = '#ppdPk-Ej1Yeb-LgbsSe-tJiF1e > img'

# 여러 페이지(999)에서 반복하기
for i in range(90):
    # 시간 지연
    time.sleep(3)
    
# 컨테이너(가게) 데이터 수집 // div.section-result-content
    stores = driver.find_elements_by_css_selector(locates)
#     print(stores)
    for s in stores:
        # 가게 이름 데이터 수집
        print(s.text)

    # 다음페이지 버튼 클릭 하기
    # 다음페이지가 없는 경우(데이터 수집 완료) 에러 처리
    try:
        nextpage = driver.find_element_by_css_selector(button)
        nextpage.click()
    except:
        break

# 크롬창 닫기
# driver.close()
```
